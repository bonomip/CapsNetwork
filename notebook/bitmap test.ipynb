```python
import cv2
import base64
import numpy as np
import io
import PIL 
from PIL import Image
from matplotlib import pyplot as plt
```

```python
pil_im = PIL.Image.open("image.png")
img = np.array(pil_im) 

print(img.shape)

im_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#making ratio 1:1
h, w = im_gray.shape
if w > h:
    d = int( (w - h) / 2)
    im_gray_crop = im_gray[0:h, d:(w-d)]
if w == h:
    im_gray_crop = im_gray
if h > w:
    d = int( (h - w) / 2)
    im_gray_crop = im_gray[d:(h-d), 0:w]

#size it down to (28, 28)
im = cv2.resize(im_gray_crop, (28, 28), interpolation = cv2.INTER_AREA)

#negativize
im = ( 255 - im )

im2 = (im > 255*0.4) * im # to save as reference

#normalize
im = im / 255.0
#threshold
im = (im > 0.4) * im




```
